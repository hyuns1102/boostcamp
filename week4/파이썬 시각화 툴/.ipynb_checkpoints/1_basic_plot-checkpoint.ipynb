## matplotlib

- pyplot 객체를 사용하여 데이터를 표시
- Pyplot 객체에 그래프들을 쌓은 다음 show로 flush
- 최대 단점 argument를 kwargs 받음
- 고정된 argument가 없어서 shift-tab이 적용이 안됨

- plot : 그림을 그려준다.

```python
import matplotlib.pyplot as plt

X = range(100)
Y = range(100)
# plt.plot(X, Y)
```

```python
plt.show()
```

```python
import numpy as np 

X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.plot(Y_2, X_2)
plt.plot(range(100), range(100))  # plot : 그림을 그려준다.
```

```python
plt.show() # pyplot 에 flush 되어서 나옴
```

```python
fig = plt.figure() # figure라는 그림판 반환
fig.set_size_inches(10,10) # 전체 그림판의 크기지정
ax_1 = fig.add_subplot(2,2,1) # 1개의 plot 생성 (그림판 안에 다른 그림판을 그린다.)
ax_2 = fig.add_subplot(2,2,2)  # 2개의 plot 생성
ax_3 = fig.add_subplot(2,2,3)  # 3개의 plot 생성
ax_4 = fig.add_subplot(2,2,4)  # 4개의 plot 생성
# plot 이 총 2행 2열 이므로, 4개 까지 들어갈 수 있다.

ax_1.plot(X_1, Y_1, c="b")  # 첫번째 plot
ax_2.plot(X_2, Y_2, c="g")  # 두번째 plot
ax_3.plot(X_1, Y_1, c="c")  # 세번째 plot
ax_4.plot(X_2, Y_2, c="y")  # 네번째 plot
plt.show() # show & flush
```

```python
plt.show()
```

### Set color
- http://matplotlib.org/2.0.2/api/colors_api.html
- Color 속성을 사용
- Float -> 흑백, rgb color, predefined color 사용
- 검색을 이용하자.

```python
X_1 = range(100)
Y_1 = [value for value in X]

X_2 = range(100)
Y_2 = [value + 100 for value in X]

plt.plot(X_1, Y_1, color="#000000")
plt.plot(X_2, Y_2, c="c")

plt.show()
```

### Set linestyle
- https://matplotlib.org/examples/lines_bars_and_markers/linestyles.html
- ls, linestyle
- 검색을 이용하자.

```python
plt.plot(X_1, Y_1, c="b", linestyle="dashed")
plt.plot(X_2, Y_2, c="r", ls="dotted")

plt.show()
```

## Set Title

- Latex 문법 가능

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed")
plt.plot(X_2, Y_2, color="r", linestyle="dotted")

plt.title("Two lines")
plt.show()
```

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed")
plt.plot(X_2, Y_2, color="r", linestyle="dotted")

plt.title('$y = \\frac{ax + b}{test}$')
plt.show()
```

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed")
plt.plot(X_2, Y_2, color="r", linestyle="dotted")

plt.title('$y = ax+b$')
plt.xlabel('$x_line$')
plt.ylabel('y_line')

plt.show()
```

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed")
plt.plot(X_2, Y_2, color="r", linestyle="dotted")

plt.text(50, 70, "Line_1")
plt.annotate(
    'line_2', xy=(50, 150), xytext=(20, 175),
    arrowprops=dict(facecolor='black', shrink=0.05))

plt.title('$y = ax+b$')
plt.xlabel('$x_line$')
plt.ylabel('y_line')

plt.show()
```

## Set legend

- Legend 함수로 범례를 표시함, loc 위치등 속성 지정
- label을 이용해, x, y title 지정 가능

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed", label='line_1')
plt.plot(X_2, Y_2, color="r", linestyle="dotted", label='line_2')
plt.legend(shadow=True, fancybox=False, loc="upper right")

plt.title('$y = ax+b$')
plt.xlabel('$x_line$')
plt.ylabel('y_line')


plt.show()
```

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed", label='line_1')
plt.plot(X_2, Y_2, color="r", linestyle="dotted", label='line_2')
plt.legend(shadow=True, fancybox=True, loc="lower right")


plt.grid(True, lw=0.4, ls="--", c=".90") # grid
plt.xlim(-1000, 2000)  # 범위 한계
plt.ylim(-1000, 2000)

plt.show()
```

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed", label='line_1')
plt.plot(X_2, Y_2, color="r", linestyle="dotted", label='line_2')

plt.grid(True, lw=0.4, ls="--", c=".90")
plt.legend(shadow=True, fancybox=True, loc="lower right")
plt.xlim(-100, 200)
plt.ylim(-200, 200)
plt.savefig("test.png")
plt.show()
```

### Scatter

- marker : scatter 모양 지정

```python
data_1 = np.random.rand(512, 2)
data_2 = np.random.rand(512, 2)
```

```python
plt.scatter(data_1[:,0], data_1[:,1], c="b", marker="x")
plt.scatter(data_2[:,0], data_2[:,1], c="r", marker="o")

plt.show()
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2 
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
# s: 데이터의 크기를 지정, 데이터의 크기 비교 가능
plt.show()
```

### Bar chart

- 조건이 까다롭기 때문에 필요할 때 찾아보자.

```python
data = [[5., 25., 50., 20.],
        [4., 23., 51., 17],
        [6., 22., 52., 19]] # y의 값

X = np.arange(0,8,2)
```

```python
plt.bar(X + 0.00, data[0], color = 'b', width = 0.50) # 브로드 캐스팅이 일어남
plt.bar(X + 0.5, data[1], color = 'g', width = 0.50) # width 만큼 X의 값을 더함
plt.bar(X + 1.0, data[2], color = 'r', width = 0.50)
plt.xticks(X+0.50, ("A","B","C", "D"))
plt.show()
```

```python
data = np.array([[5., 25., 50., 20.],
        [4., 23., 51., 17],
        [6., 22., 52., 19]])

color_list = ['b', 'g', 'r']
data_label = ["A","B","C"]
X = np.arange(data.shape[1])
X
```

```python
data = np.array([[5., 5., 5., 5.],
        [4., 23., 51., 17],
        [6., 22., 52., 19]])

for i in range(3):
    plt.bar(X, data[i], bottom = np.sum(data[:i], axis=0), 
            color = color_list[i], label=data_label[i])
plt.legend()
plt.show()

```

```python
A = [5., 30., 45., 22.]
B = [5, 25, 50, 20]

X = range(4)

plt.bar(X, A, color = 'b')
plt.bar(X, B, color = 'r', bottom = 60)
plt.show()
```

```python
women_pop = np.array([5, 30, 45, 22])
men_pop = np.array([5, 25, 50, 20])
X = np.arange(4)

plt.barh(X, women_pop, color = 'r')
plt.barh(X, -men_pop, color = 'b')
plt.show()
```

## Histogram

```python
X = np.arange(1000)
# plt.hist(X,bins=100)

Y = np.random.randn(1000)
# plt.hist(Y,bins=100)

fig = plt.figure()  # pyplot 안에 figure 그림판 정의
fig.set_size_inches(10, 10)  # 그림판 사이즈 정의
ax1 = fig.add_subplot(2,1,1) # 그림판 안에 또다른 plot 정의
ax2 = fig.add_subplot(2,1,2)

ax1.hist(X, bins =100)  # 정의된 plot 안에 histogram 그려주기
ax2.hist(Y, bins =100)

plt.show()
```
